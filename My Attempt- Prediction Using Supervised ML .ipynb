### What is Linear Regression
**Linear regression analysis** predicts the value of one variable based on the value of another.
The variable to be predicted is known as the *dependent variable*. The score, for example, is the dependent variable in this case.

The variable we use to predict the value of the other variable is known as the *independent variable*. The independent variable here is the number of study hours.

*Linear regression is used to fit a straight line or surface that minimises the differences between predicted and actual output values.*
